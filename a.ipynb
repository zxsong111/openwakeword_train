```python


import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
# 1. 
y, sr = librosa.load("dataset/hey_aldelo/positive_train/0a9a5903-0a30-4d2e-b6b6-891f73d4b4e0_0e5ea66e-7ce9-420f-a101-34073d973e27.wav", sr=16000)

# 2.  Mel-spectrogram
mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=1024, hop_length=256, n_mels=80)

# 3. 
log_mel_spec = librosa.power_to_db(mel_spec, ref=np.max)

# 4. 
plt.figure(figsize=(10, 4))
librosa.display.specshow(log_mel_spec, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Log-Mel Spectrogram')
plt.tight_layout()
plt.show()
```
